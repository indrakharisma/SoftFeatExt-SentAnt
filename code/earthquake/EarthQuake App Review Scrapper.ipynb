```python
result1 = reviews_all(
    'com.jrustonapps.myearthquakealerts',
    sleep_milliseconds=1, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
    filter_score_with=None # defaults to None(means all score)
)
x=json.dumps(result1, indent=4, sort_keys=True, default=str)
with open('xx.txt', 'w', encoding='utf-8') as f:
    f.write(x)
my_json = re.escape(x).replace('\\',"")
my_json = my_json.replace('"bonnnng"',"bonnnng")
my_json = my_json.replace("'ve"," have")
my_json = my_json.replace('"Report problem with app"',"Report problem with app")
my_json = my_json.replace('"Show Magnitude"',"Show Magnitude")
my_json = my_json.replace('"Yes, I felt that"',"Yes, I felt that")
my_json = my_json.replace('"near me"',"near me")
my_json = my_json.replace('"normal"',"normal")
my_json = my_json.replace('"1"',"1")
my_json = my_json.replace('"Show Region"',"Show Region")
my_json = my_json.replace('"Notifications"',"Notifications")
my_json = my_json.replace('"Recent"',"Recent")
my_json = my_json.replace('"Notification Problems"',"Notification Problems")
my_json = my_json.replace('"movement"',"movement")
my_json = my_json.replace('"My Region"',"My Region")
my_json = my_json.replace('"Unit of Measurement"',"Unit of Measurement")


print(my_json)

data = json.loads(my_json)
#ambil atribut content aja
txtReview = [data['content'] for data in data]

#convert JSON ke String
stringReview = ".\n".join(filter(None, txtReview))
print(stringReview)

with open('1-Ulasan-Aplikasi-myearthquakealerts.txt', 'w', encoding='utf-8') as f:
    f.write(stringReview)
    
```
